## Data Understanding

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('googleplaystore.csv')
df.head()
print("Shape:", df.shape)
print(df.info())
print(df.isnull().sum())
```

## Data Cleaning

```python

df = df.drop_duplicates()

df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce')

df['Installs'] = df['Installs'].astype(str).str.replace('+','', regex=False).str.replace(',','', regex=False)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')

df['Price'] = df['Price'].astype(str).str.replace('$','', regex=False)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

def convert_size(x):
    try:
        x = str(x)
        if 'M' in x:
            return float(x.replace('M',''))
        elif 'k' in x:
            return float(x.replace('k',''))/1024
        else:
            return np.nan
    except:
        return np.nan

df['Size_MB'] = df['Size'].apply(convert_size)

df['Rating'].fillna(df['Rating'].median(), inplace=True)
```

## Univariate Analysis

```python

sns.histplot(df['Rating'], bins=20)
plt.title('Rating Distribution')
plt.show()
```


### Observation
Most applications have ratings between 4.0 and 4.5, indicating generally positive user feedback.


```python

plt.figure(figsize=(12,5))
df['Category'].value_counts().plot(kind='bar')
plt.title('Apps by Category')
plt.show()
```

```python

df['Type'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.ylabel('')
plt.show()
```

## Bivariate Analysis

```python

sns.scatterplot(data=df, x='Reviews', y='Rating')
plt.show()
```

```python

plt.figure(figsize=(14,5))
sns.boxplot(data=df, x='Category', y='Rating')
plt.xticks(rotation=90)
plt.show()
```

```python

category_installs = df.groupby('Category')['Installs'].mean().sort_values(ascending=False)
category_installs.head(10)
```

## Correlation Analysis

```python

num_df = df[['Rating','Reviews','Installs','Price']]
sns.heatmap(num_df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.show()
```


# Key Insights

1. Most Play Store apps are free.
2. Ratings are concentrated between 4.0–4.5.
3. Family, Game and Tools categories contain the highest number of apps.
4. Reviews and installs show a strong positive relationship.
5. Paid apps form a very small percentage of the dataset.
6. Popular apps tend to accumulate both installs and reviews.
7. Price has little impact on ratings.



# Conclusion

The Google Play Store dataset contains over 10,000 applications across multiple categories. Most apps are free and receive ratings above 4.0. Family and Game categories dominate the marketplace. Reviews and installs are strongly related, suggesting user engagement drives popularity. Overall, app success appears more dependent on visibility and engagement than on price.
